```python
import random

# Opciones posibles
opciones = ["piedra", "papel", "tijeras", "lagarto", "spock"]

# Contadores
victorias_usuario = 0
victorias_computadora = 0

# FunciÃ³n para pedir el nombre del jugador
def pedir_nombre():
    nombre = input("Antes de humillarte, dime... Â¿cÃ³mo te llamas? â†’ ").strip()
    if not nombre:
        nombre = "Humano sin nombre"
    return nombre

nombre_usuario = pedir_nombre()

print(f"\nSaludos, {nombre_usuario} ðŸ¤–\nHoy tendrÃ¡s el honor â€”y la humillaciÃ³n estadÃ­sticamente inevitableâ€” de enfrentarte al mayor genio en Piedra, Papel, Tijeras, Lagarto, Spock. \nPrepÃ¡rate para una derrota elegante.ðŸ’¥\nElige tu arma")

# FunciÃ³n para jugar una ronda
def jugar_ronda():
    usuario = input("\nElige piedra, papel, tijeras, lagarto o spock â†’ ").lower()
    computadora = random.choice(opciones)

    print(f"Tu elecciÃ³n fue: {usuario}")
    print(f"El gran Sheldon eligiÃ³: {computadora}")

    if usuario == computadora:
        print("Â¡Empataste con el maestro!")
        return 0  # empate
    elif (
        (usuario == "tijeras" and computadora in ["papel", "lagarto"]) or
        (usuario == "papel" and computadora in ["piedra", "spock"]) or
        (usuario == "piedra" and computadora in ["lagarto", "tijeras"]) or
        (usuario == "lagarto" and computadora in ["spock", "papel"]) or
        (usuario == "spock" and computadora in ["tijeras", "piedra"])
    ):
        print(f"Â¡Inconcebible, {nombre_usuario}! Has derrotado al maestroâ€¦ esto requiere revisiÃ³n de datos. ðŸŽ‰")
        return 1  # gana usuario
    else:
        print(f"Como predije estadÃ­sticamente, {nombre_usuario}, tu derrota era inevitable... Bazinga!. ðŸ˜¢")
        return -1  # gana computadora

# Bucle principal para mejor de 3
while victorias_usuario < 2 and victorias_computadora < 2:
    resultado = jugar_ronda()
    if resultado == 1:
        victorias_usuario += 1
    elif resultado == -1:
        victorias_computadora += 1

    print(f"Marcador â†’ {nombre_usuario}: {victorias_usuario} | Sheldon: {victorias_computadora}")

# Resultado final
if victorias_usuario > victorias_computadora:
    print(f"\nContra toda probabilidad... Â¡has ganado la serie al mejor de 3, {nombre_usuario}!")
else:
    print(f"\nEl universo sigue en orden: Sheldon ha ganado la serie al mejor de 3 contra {nombre_usuario}.")
```
